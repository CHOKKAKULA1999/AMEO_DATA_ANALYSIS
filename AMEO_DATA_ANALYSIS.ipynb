HERE I AM CALCULATING THE OUTLIERS FOR EACH NUMERICAL COLUMN SO THAT EVERY NUMERICAL COLUMN WILL BE CHECKED AND THE COLUMN WITH MOST OUTLIERS WILL BE SHOWED FIRST THE NUMBERS OF OUTLIERS IN THAT PARTICULAR COLUMN 

```python

# Select the numerical columns
num_cols = ['Salary', '10percentage', '12graduation', '12percentage', 'collegeGPA', 'English', 'Logical', 'Quant', 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']

# Calculate the number of outliers for each numerical column
num_outliers = {}
for col in num_cols:
    q1 = np.percentile(AMEO_DF[col], 25)
    q3 = np.percentile(AMEO_DF[col], 75)
    iqr = q3 - q1
    upper = q3 + 1.5*iqr
    lower = q1 - 1.5*iqr
    num_outliers[col] = len(AMEO_DF[(AMEO_DF[col] > upper) | (AMEO_DF[col] < lower)])

# Sort the dictionary by the number of outliers in descending order
sorted_outliers = {k: v for k, v in sorted(num_outliers.items(), key=lambda item: item[1], reverse=True)}

# Print the column with the most outliers and the number of outliers
print("Column with the most outliers:", list(sorted_outliers.keys())[0])
print("Number of outliers:", list(sorted_outliers.values())[0])
```
