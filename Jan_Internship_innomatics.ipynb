#MANDATORY OPERATIONS

**1)Group by rating based on movieId and apply aggregations like count and mean**

```python
import pandas as pd

# Read the CSV file into a DataFrame
ratings_df = pd.read_csv('ratings.csv')

# Group by movieId and apply aggregations
grouped_ratings = ratings_df.groupby('movieId').agg(count=('userId', 'count'), mean_rating=('rating', 'mean'))

# Print the results
print(grouped_ratings)
```

**2)Apply inner join on dataframe created from ratings.csv and movies.csv**

```python
# Perform an inner join on the DataFrames
joined_df = pd.merge(grouped_ratings, movies_df, on='movieId', how='inner')

# Print the resulting DataFrame
print(joined_df)
```

**3)Filter only those movies with more than 50 ratings**

```python
# Filter movies with more than 50 ratings
filtered_movies = joined_df[joined_df['count'] > 50]

# Print the filtered movies
print(filtered_movies)
```
