I SEE SALARY COLUMN HAS MANY OUTLIERS AND I WANT TO REMOVE THOSE 

```python
Q1 = AMEO_DF['Salary'].quantile(0.25)
Q3= AMEO_DF['Salary'].quantile(0.75)

IQR = Q3- Q1
AMEO_DF = AMEO_DF[(AMEO_DF['Salary'] >= Q1 -1.5* IQR) &(AMEO_DF['Salary'] <= Q3 + 1.5*IQR)]
```

**AN INTRESTING TEST I JUST GOT IS CHI_SQUARE_TEST IT TELL THE TEST OF INDEPENDENCE OF TWO COLUMNS **

#THE NULL HYPOTHESIS ASSUMES THAT THEY ARE INDEPENEDENT . IF THE P-VALUE IS LESS THAN SIGNIFICANCE LEVEL , WE REJECT THE NULL HYPOTHESIS AND CONCLUDE THAT THE VARIABLE ARE DEPENDENT 

DEGREE VS SPECIALIZATION

```python
import scipy.stats as stats
# creating the contigency table 
con_table = pd.crosstab(AMEO_DF['Degree'] ,AMEO_DF['Specialization'])

#performing chi-square test
chi2 , p , dof ,expected =stats.chi2_contingency(con_table)

#printing the results 
print(con_table)
print("chi2_square statistics:" ,chi2)
print("p-value", p)
print("Degree of freedom:" , dof)
print("Expected frequency:",expected)
```

WE CAN SEE THAT P-VALUE IS 2.38 WHICH IS GREATER THAN 0.05 THE SIGNIFICANCE VALUE SO WE ACCEPT THE NULL HYPOTHESIS AND TWO VARIABLE ARE DEPENDENT WITH EACH OTHER
